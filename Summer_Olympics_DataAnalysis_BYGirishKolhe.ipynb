# Summer Olympics Data Analysis Assignment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("summer.csv")
df.head()
```

### 1. In how many cities Summer Olympics is held so far?

```python
oly_city = df['City'].unique()
l =[]
for i in oly_city:
    l.append(i)
print("Cities were olympics held: ",l)
print("\nTotal number of cities: ",len(oly_city))
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
goldMedals = []
g1 = df[df['Medal'] == 'Gold']
for i in g1['Sport'].unique():
    goldMedals.append([i,len(g1[g1['Sport']==i])])
g2 = pd.DataFrame(goldMedals,columns = ['Sports','Gold']).sort_values(by='Gold', ascending=False).head()
g2.plot(x = 'Sports', y = 'Gold', kind = 'bar', figsize = (7,5),color="orange",title="Sports is having most number of Gold Medals")
g2L = g2.values
print("{} Sport has Won {} Number of Gold Medals.".format(g2L[0][0],g2L[0][1]))
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
allMedals=[]
for m in df['Sport'].unique():
    allMedals.append([m,len(df[df['Sport'] == m])])
m1 = pd.DataFrame(allMedals,columns = ['Sports','Medals']).sort_values(by='Medals', ascending=False).head()
m1.plot(x = 'Sports',y='Medals',color='green',kind='bar',figsize = [7,5],title='Sports is having most number of medals')
print("{} Sport has won {} Medals(Gold, Silver, Bronze)".format(m1['Sports'][0],m1['Medals'][0]))
```

### 4. Which player has won most number of medals? (Top 5)

```python
plyName =[]
for p1 in df['Athlete'].unique():
    plyName.append([p1,len(df[df['Athlete'] == p1])])
p2 = pd.DataFrame(plyName,columns = ['Athletes','Medals']).sort_values(by='Medals', ascending=False).head()
p2.plot(x = 'Athletes',y='Medals',kind='bar',figsize = [7,5],title="Player Winning Most Medals in all Olympics")
p2L = p2.values
print("{} has won {} number of Medals(Gold, Silver, Bronze)".format(p2L[0][0],p2L[0][1]))
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
pgName =[]
pg1 = df[df['Medal'] == 'Gold']
for plgold in pg1['Athlete'].unique():
    pgName.append([plgold,len(pg1[pg1['Athlete']==plgold])])
pg2 = pd.DataFrame(pgName,columns = ['Athletes','GoldMedals']).sort_values(by='GoldMedals', ascending=False).head()
pg2.plot(x = 'Athletes',y='GoldMedals',kind='bar',figsize = [7,5],title="Player Winning Most Gold Medals in all Olympics",color='orange')
pg2L=pg2.values
print("{} has won {} number of Gold Medals".format(pg2L[0][0],pg2L[0][1]))
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
cou = df[df['Country'] == "IND"]
couG = cou[cou['Medal']=='Gold']
y = couG['Year'].unique()
spo = couG['Sport'].unique()
print("India Won first Olympic Gold Medal in {} at Year {} ".format(spo[0],y[0]))
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
events=[]
for e in df['Event'].unique():
    events.append([e,len(df[df['Event'] == e])])
eve1 = pd.DataFrame(events,columns = ['Athletes','Events']).sort_values(by='Events', ascending=False).head()
eve1.plot(x = 'Athletes',y='Events',kind='bar',figsize = [7,5],title="Most Event played by player in all Olympics")
eveL = eve1.values #converting it to list
print("{} is the event were played by {} players.".format(eveL[0][0],eveL[0][1]))
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
femaleSport=[]
pg1 = df[df['Medal'] == 'Gold']
pfg = pg1[pg1['Gender'] == 'Women']
for w in pfg['Sport'].unique():
    femaleSport.append([w,len(pfg[pfg['Sport'] == w])])
pfg2 = pd.DataFrame(femaleSport,columns = ['Sports','GoldMedals']).sort_values(by='GoldMedals', ascending=False).head()
pfg2.plot(x = 'Sports',y='GoldMedals',kind='bar',color='yellow',figsize = [7,5],title="Sports were Female Players Winning Most Gold Medals in all Olympics")
pfg2L=pfg2.values
print("{} Sport were female has won {} Gold Medals".format(pfg2L[0][0],pfg2L[0][1]))
```
